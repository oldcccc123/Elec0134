```python
import pandas as pd
from skimage import io
import os
```

```python
file = './dataset/image/' 
csv='./dataset/label.csv'
data = pd.read_csv(csv)
for n in range(len(data)):
    image = io.imread(file + str(data.iloc[n, 0]), as_gray=True)
    if not os.path.exists('./dataset/' + data.iloc[n, 1]+'/'):
        os.makedirs('./dataset/' + data.iloc[n, 1]+'/')
    io.imsave('./dataset/' + data.iloc[n, 1]+'/'+str(data.iloc[n, 0]),image)
```
